## subplots 함수를 이용하여 axes 여러개 그리기
- 함수를 실행하면 figure, axes 객체가 반환된다.


````python
fig, ax = plt.subplots(2,2,figsize=[15,5]) # 2행 2열
print(ax.shape)
ax[0,0].plot(np.arange(5))

ax[0,1].plot(np.arange(2,6))
ax[0,1].plot(np.arange(5))

ax[1,0].plot(np.arange(5))

ax[1,1].plot(np.arange(5))
plt.show()
````

- 다른 방식으로 여러개 그리기

## 타이틀 설정하기
- figure 객체의 제목은 subtitle 메소드
- axes 객체의 제목은 set_title 메소드

````python
fig ,ax = plt.subplots(1,2,figsize=(15,5))
fig.suptitle("figure title")

x = range(0,10) # x축
y = np.exp(x) # y축
ax[0].plot(x,y)
ax[0].set_title("exp")

x = range(1,1000) # x축
y = np.log(x) # y축
ax[1].plot(x,y)
ax[1].set_title("log")
plt.show()
````

## x와 y 축에 이름 넣기

````python
x = range(1,1000)
y = np.log(x)
````

## Markers, line style, color 변경하기
- plot 함수에 아규먼트를 지정해주면 된다.


````python
y = np.exp(range(0,10))
fig , ax = plt.subplots(2,2,figsize=[15,10])
ax[0,0].plot(y,marker="o") 
# 원

ax[0,1].plot(y,marker="v",linestyle='--',color="y") 
# 역삼각형 , -- 선, 노란색

ax[1,0].plot(y,marker="x",linestyle="",color="r") 
# x , 공백을 주면 라인은 없음, 빨강색

ax[1,1].plot(y,marker="^",color="g") 
# 삼각형 , 초록색
plt.show()
````

## 범례 표시하기

````python
fig, ax = plt.subplots()
x = np.arange(10)
ax.plot(x)
ax.plot(x**2,alpha = 0.3) # alpha 투명도 0~1 사이값을 넣어주면 된다.
ax.legend(["X","X^2"],loc=[0.5,0.8]) # 각각의 선에 범례를 지정해줌 loc는 범례박스 위치 지정
plt.show()
````

# Box Plot
- 4분위수를 기준으로 데이터의 최대, 최소, Q1(25%지점), Q2(50%지점), Q3(75%지점)등 통계적인 부분들을 손쉽게 시각화 
- 이상치도 쉽게 파악할 수 있다.
- IQR(Inter Quantile Range): 
    - Q3 - Q1 
    - 3사분위수와 1사분위수의 차이
- Box Plot 최대값 : Q3 + IQR * 1.5
- Box Plot 최소값 : Q1 - IQR * 1.5
- Box Plot 에서는 최소값과 최대값을 벗어나면 이상치로 판단한다.


````python
np.arange(0,1.1,0.25) # 사분위수
````

````python
q_arr = df["팁금액"].quantile(np.arange(0,1.1,0.25)).to_numpy()
q_arr
````

````python
iqr = q_arr[3] - q_arr[1]
iqr
box_min = q_arr[1] - iqr * 1.5 # 박스플롯의 최소값
box_max = q_arr[3] + iqr * 1.5
iqr,box_min,box_max
````

````python
yticks = np.concatenate([
    q_arr, (iqr,box_min,box_max)
])
yticks
````

````python
fig,ax = plt.subplots(figsize = (15,10))
ax.boxplot(df["팁금액"])
ax.set_yticks(yticks)
plt.show() #데이터 최소값이 1이여서 1에 배치 높은값에 이상치가 있다.
````
