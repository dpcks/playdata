- 다중 조건을 주어 마스킹
    - 데이터프레임에서는 논리 연산자를 지원하지 않는다.
    - `or` 대신 `|`(Vertical bar)
        - 둘중에 하나가 1 이면 1을 반환
    - `and` 대신 `&`(Ampersand)
        - 둘다 1이면 1을 반환
    -  `not` 대신 `~`(tilde)
        ` 1이면 0을 반환

```python
True == 1 , False == 0
```

```python
mask1 = df2["target"] == 1 
mask2 = df2["age"] < 20
df2.loc[mask1 & mask2].head()
```

```python
mask1 = df2["target"] == 1
mask2 = df2["embarked"] == "S" 
df2.loc[mask1 & mask2].head()
```

- map 메소드 사용해보기
    - 딕셔너리를 전달해서 컬럼의 값들을 변경할 수 있다.
    - `key`는 변경전 값, `value`는 변경후 값

```python
replace_dict = {"male":1 ,"female":0} # 전체데이터중 비율이 적은 쪽을 1로 줌
df["gender"].map(replace_dict)
```
